```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

# Load the Iris dataset
iris = pd.read_csv('iris.csv')

# Display the first few rows of the dataset
print(iris.head())

```

```python
# Univariate Analysis
# Plotting histograms and KDE plots for each feature
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
sns.histplot(iris['sepal_length'], kde=True)
plt.title('Sepal Length Distribution')

plt.subplot(2, 2, 2)
sns.histplot(iris['sepal_width'], kde=True)
plt.title('Sepal Width Distribution')

plt.subplot(2, 2, 3)
sns.histplot(iris['petal_length'], kde=True)
plt.title('Petal Length Distribution')

plt.subplot(2, 2, 4)
sns.histplot(iris['petal_width'], kde=True)
plt.title('Petal Width Distribution')

plt.tight_layout()
plt.show()
```

```python

# Bivariate Analysis
# Pairwise relationships in the dataset using pairplot
sns.pairplot(iris, hue='species', palette='Set1')
plt.show()

```

```python

# Multivariate Analysis
# Violin plots to see the distribution of features with respect to species
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
sns.violinplot(x='species', y='sepal_length', data=iris, palette='Set2')
plt.title('Sepal Length by Species')

plt.subplot(2, 2, 2)
sns.violinplot(x='species', y='sepal_width', data=iris, palette='Set2')
plt.title('Sepal Width by Species')

plt.subplot(2, 2, 3)
sns.violinplot(x='species', y='petal_length', data=iris, palette='Set2')
plt.title('Petal Length by Species')

plt.subplot(2, 2, 4)
sns.violinplot(x='species', y='petal_width', data=iris, palette='Set2')
plt.title('Petal Width by Species')

plt.tight_layout()
plt.show()

# Pairplot with species hue
sns.pairplot(iris, hue='species', diag_kind='kde', palette='Set1')
plt.show()
```
